```python
# importing libraries
import numpy as np
import pandas as pd
```

```python
# importing data
df = pd.read_csv("startup_funding.csv")
df.head()
```

```python
#List of startups
sorted(df["Startup Name"].unique().tolist())
```

```python
df.info()
```

```python
#Filling the null values
df["Investors Name"] = df["Investors Name"].fillna("Undisclosed")
```

```python
# List of investors
sorted(df["Investors Name"].unique().tolist())
```


# Data Cleaning

```python
# droping the unwanted columns
df.drop(["Remarks"],axis = 1,inplace = True)
```

```python
# setting the Sr.no as index
df.set_index(["Sr No"],inplace = True)
```

```python
# Renameing the column names in dataframe
df.rename(columns = {
    'Date dd/mm/yyyy' : 'date',
    "Startup Name" : 'startup',
    "Industry Vertical" : 'vertical',
    'SubVertical' : 'subvertical',
    'City Location' : 'city',
    'Investors Name' : 'Investors',
    'InvestmentnType' : 'round',
    'Amount in USD' : 'amount'
},inplace = True)
```

```python
# Filling the null values with zero
df["amount"] = df["amount"].fillna('0')
```

```python
df["amount"] = df["amount"].str.replace(",","")
df["amount"] = df["amount"].str.replace("undisclosed","0")
df["amount"] = df["amount"].str.replace("Undisclosed","0")
df["amount"] = df["amount"].str.replace("unknown","0")
```

```python
# Filtering the data with amount column
df = df[df["amount"].str.isdigit()]
```

```python
# Changing the datatype of amount from str to int
df["amount"] = df["amount"].astype('float')
```

```python
df.head()
```

```python
# Function to convert dollors into cr(inr)
def to_inr(dollar):
    inr = dollar * 82.5
    return inr/10000000
```

```python
# converting amount from dollors to inr
df["amount"] = df["amount"].apply(to_inr)
```

```python
df.head()
```

```python
df['date'] = df['date'].str.replace("05/072018","05/07/2018")
```

```python
# converting date column object to datatimeobject
df['date'] = pd.to_datetime(df['date'],errors="coerce")
```

```python
# Extacting month from date
df["date"].dt.month
```

```python
df.info()
```

```python
# Cleaning the null values
df = df.dropna(subset = ['date','vertical','subvertical','City  Location','round','amount'])
```

```python
# Extracting the cleaned dataset for web page
df.to_csv("startup_cleaned.csv",index = False)
```
